```python
data = pd.read_csv('multipliers.csv', thousands=',', nrows=105)
data.tail()
```

```python
# convert wide table to long table
multipliers = pd.melt(data, 
                     id_vars='product', 
                     value_vars=data.columns.to_list()[1:]).dropna()

multipliers.shape
```

```python
# Strip whitespace from start and finish

def space_stripper(x):
    return x.strip()

multipliers[['product','variable']] = multipliers[['product','variable']].applymap(lambda x: space_stripper(x))
multipliers.head()
```

```python
def dotter(x):
    if x=='.' or x=='..':
        return np.nan
    else:
        return x

multipliers = multipliers.applymap(lambda x: dotter(x))
multipliers.head()
```

```python
multipliers['value'] = multipliers['value'].astype(float)
multipliers.head()
```

```python
multipliers['factor'] = 'Ratio of direct to total'
multipliers.head()
```
