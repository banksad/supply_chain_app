```python
import numpy as np
import pandas as pd
import os

os.chdir('..')
```

```python
data = pd.read_excel('data/nasu1719pr.xlsx',sheet_name='Imports use pxi',header=4)
data = data.iloc[1:105,1:106]
data = data.rename(columns={data.columns[0]: 'product'})
data
```

```python
cpa = pd.read_excel('data/nasu1719pr.xlsx',sheet_name='Use BP PxI',header=4)
cpa = pd.DataFrame(cpa.iloc[1:105,0])
cpa = cpa.rename(columns={cpa.columns[0]: 'CPA'})
cpa
```

```python
sic = pd.read_excel('data/nasu1719pr.xlsx',sheet_name='Use BP PxI',header=3)
sic = pd.DataFrame(sic.columns[1:106])
sic = sic.rename(columns={sic.columns[0]: 'SIC'})
sic
```

```python
data['product'] = cpa['CPA'].astype(str).str.cat(data['product'].astype(str), sep=' ')
data['product'] = data['product'].apply(lambda x: x[4:])
data
```

```python
data = data.T
data.index = sic['SIC'].astype(str).str.cat(data.index.astype(str), sep=' ')
data = data.T
data = data.rename(columns={data.columns[0]: 'product'})
data
```

```python
# convert wide table to long table
data_long = pd.melt(data, 
                     id_vars='product', 
                     value_vars=data.columns.to_list()[1:]).dropna()

data_long.head()
```

```python
# Strip whitespace from start and finish

def space_stripper(x):
    return x.strip()

data_long[['product','SIC']] = data_long[['product','SIC']].applymap(lambda x: space_stripper(x))
```

```python
data_long.rename(columns={'product':'input requirements'}, inplace=True)
data_long.rename(columns={'SIC':'industry'}, inplace=True)
data_long
```

```python
data_long['proportion'] = data_long['value'] / data_long.groupby('industry')['value'].transform('sum')

data_long.to_csv('data/imports_pi.csv')
```
