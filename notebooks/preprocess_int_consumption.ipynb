```python
import numpy as np
import pandas as pd
import os

os.chdir('..')
```

```python
data = pd.read_excel('data/nasu1719pr.xlsx',sheet_name='Use BP PxI',header=4)
data = data.iloc[1:106,1:107]
data = data.rename(columns={data.columns[0]: 'product'})
data.head()
```

Examine imports use table

```python
# convert wide table to long table
data_long = pd.melt(data, 
                     id_vars='product', 
                     value_vars=data.columns.to_list()[1:]).dropna()

data_long.shape
data_long.head()
```

```python
# Strip whitespace from start and finish

def space_stripper(x):
    return x.strip()

data_long[['product','variable']] = data_long[['product','variable']].applymap(lambda x: space_stripper(x))
```

```python
data_long.rename(columns={'product':'input requirements'}, inplace=True)
data_long.rename(columns={'variable':'industry'}, inplace=True)
```

```python
data_long['proportion'] = data_long['value'] / data_long.groupby('industry')['value'].transform('sum')

data_long.to_csv('industry_use_cleaned.csv')
```
